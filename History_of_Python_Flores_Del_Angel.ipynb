**WHAT IS PYTHON AND WHAT MAKES IT SO INTERESTING?**

Python is an interpreted programming language whose main philosophy is that it be readable by anyone with basic programming knowledge. *   It is totally free. It is an open source or open source language, so there is no license fee to use it.

* It is backed by a huge community. Its free nature means that new libraries and applications are continually being developed. It's hard to think of something someone hasn't done. This is a multiplying factor for programmers, since any questions will be resolved in the forums.

* It is a multiparadigm language. This means that it combines properties of different programming paradigms, which allows it to be very flexible and easy to learn independently of the knowledge of the interested party.

* Its applications are not limited to a specific area. The fact that it is multiparadigm allows it to be used in seemingly disparate fields such as web application design or artificial intelligence, among many others.

* Python is suitable for all platforms. We can run it on different operating systems such as Windows or Linux simply by using the corresponding interpreter.

**PYTHON STATEMENTS**

**Print statement:**

The **print** statement in Python provides the functionality to print string and variable. The printed statement takes the message to be printed in quotes. A comma is used to print the variable along with the message. The print declaration evaluates each expression that is separated by a comma. If an expression is not a string, it will be converted to a string and then displayed. And the print statement is always followed by a new line unless it ends with a comma.

EXAMPLE:

```python
print ("Hi! My name is Marco")
```

With the print statement you can also **print values and solve operations**.

 EXAMPLE:

```python
print (10)
x = 5
print (x)
```

```python
print (10*10)
```

*Parameter end:*

Python's print () function comes with a parameter called "end". By default, the value of this parameter is "\ n", that is, the new line character. You can end a print declaration with any character / string using this parameter.

EXAMPLE:

```python
print("Welcome to" , end = ' ') 
print("Universidad Tecnologica de Altamira", end = ' ')
```

**If-else statement:**

A Python if else statement evaluates whether an expression is true or false. If a condition is true, the "if" statement is executed. Otherwise, the "else" statement is executed. Python if else statements help programmers control the flow of their programs.

EXAMPLE:

```python
age = int (input ("How old are you?"))
if age <0:
    print ("You cannot have a negative age")
elif age >= 0 and age < 18:
    print ("You are a minor")
else:
    print ("You are of legal age")
```

**Loops**

A loop is a sequence that repeats the same piece of code several times, until the condition assigned to the loop is no longer met. The most common loops are:

*For loop:*

A for loop is a loop that repeats the statement block a predetermined number of times. The block of instructions that is repeated is usually called the body of the loop, and each repetition is usually called an iteration.

EXAMPLE:


```python
print("Start")
for i in [3, 4, 5]:
    print(f"Hello. Now its {i} and its square {i ** 2}")
print ("End")
```

*While Loop:*

A while loop allows repeating the execution of a group of instructions as long as a condition is met (that is, as long as the condition has the value True).

EXAMPLE:

```python
count = 0
while (count < 30):   
    count = count + 1
    print("My name is Marco")
```
